```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('ecom.csv')
df.head()
```

```python
label_encoder = LabelEncoder()
df['target_audience_encoded'] = label_encoder.fit_transform(df['Target Audience'])
df['market_conditions_encoded'] = label_encoder.fit_transform(df['Market Conditions'])
df.head()
```

```python
col = ['Cost of Production ($)', 'Competition Level (1-5)', 'Desired Profit Margin (%)', 'target_audience_encoded', 'market_conditions_encoded']
X = df[col]
Y = df['Actual Price ($)']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

```

```python
features = {
    'Cost of Production ($)': 9800,
    'Competition Level (1-5)': 2,	
    'Desired Profit Margin (%)': 4,	
    'target_audience_encoded': 2,
    'market_conditions_encoded':2
}

f_df = pd.DataFrame([features])
pre_price = model.predict(f_df)
```
